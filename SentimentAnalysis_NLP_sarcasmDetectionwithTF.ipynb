reading a JSON file and loading its data in a LIST

```
import json

#NOTE : before running this snippet, load the Sarcasm_Headlines_Dataset_v2.json dataset in PATH : /content/sarcasmDetection/Sarcasm_Headlines_Dataset_v2.json

#f = open('/content/sarcasmDetection/Sarcasm_Headlines_Dataset_v2.json','r')
#for line in f:
  #print(line)

########## READING A JSON FILE & LOADING ITS CONTENTS IN THEIR RESPECTIVE LISTS ##########
with open('/content/sarcasmDetection/Sarcasm_Headlines_Dataset_v2.json','r') as f:
  datastore = json.load(f)
   

print(datastore)

isSarcastic = []
headline = []
articleLink = []
for item in datastore:
  #print(item)
  #print(item['headline'])
  isSarcastic.append(item['is_sarcastic'])
  #print(isSarcastic)
  headline.append(item['headline'])
  articleLink.append(item['article_link'])
print(headline)

#########################################################################################
```

```
import numpy as np

training_size = 3700

training_headlines =  headline[0:training_size]
training_labels = np.array(isSarcastic[0:training_size])
testing_headlines = headline[training_size:]
testing_labels = np.array(isSarcastic[training_size:])

#print(training_headlines)
#print(testing_headlines)
```

```
## IMPORT LIBRARIES ##

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
## TOKENIZE THE STATEMENTS ##
numWords = 10000
oovToken = "<oov>"

tokenizer = Tokenizer(num_words=numWords, oov_token=oovToken)
tokenizer.fit_on_texts(training_headlines)
#tokenizer.word_index
training_sequences = tokenizer.texts_to_sequences(training_headlines)
testing_sequences = tokenizer.texts_to_sequences(testing_headlines)

print(training_sequences)
```

```
## PADDING ##
maxLen = 120
paddingStyle = 'post'

training_seq_padded = pad_sequences(training_sequences, maxlen=maxLen, padding=paddingStyle)
testing_seq_padded = pad_sequences(testing_sequences, maxlen=maxLen, padding=paddingStyle)

print(training_seq_padded)
```

```
reverse_word_index = dict([(value, key) for (key, value) in tokenizer.word_index.items()])

def decode_review(text):
    return ' '.join([reverse_word_index.get(i, '?') for i in text])

print(decode_review(training_seq_padded[100]))
print(training_headlines[100])
```

```
import tensorflow as tf

embedding_dim = 16

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(numWords, embedding_dim, input_length=maxLen),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```
num_epochs = 10
model.fit(training_seq_padded, training_labels, epochs=num_epochs, validation_data=(testing_seq_padded, testing_labels))
```

```
e = model.layers[0]
weights = e.get_weights()[0]
print(weights.shape) # shape: (vocab_size, embedding_dim)
```

```
import io

print(reverse_word_index)

out_v = io.open('vecs.tsv', 'w', encoding='utf-8')
out_m = io.open('meta.tsv', 'w', encoding='utf-8')
for word_num in range(1, numWords):
  word = reverse_word_index[word_num]
  embeddings = weights[word_num]
  out_m.write(word + "\n")
  out_v.write('\t'.join([str(x) for x in embeddings]) + "\n")
out_v.close()
out_m.close()
```
